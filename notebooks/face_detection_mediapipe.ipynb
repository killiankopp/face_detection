# correspondance sur plusieurs encodages

```python
def find_best_match(detected_face_encoding, known_encodings_file, threshold=0.6):
    with open(known_encodings_file, 'r') as f:
        known_data = json.load(f)

    best_match = None
    min_distance = float('inf')

    for label, encodings in known_data.items():
        for encoding in encodings:
            distance = calculate_distance(detected_face_encoding, np.array(encoding))
            if distance < min_distance:
                min_distance = distance
                best_match = label

    if min_distance < threshold:
        return f"Correspondance trouvée : {best_match} (distance : {min_distance:.2f})"
    else:
        return "Aucune correspondance trouvée."
```

# Liveness

```python
import cv2
import dlib
import numpy as np
from scipy.spatial import distance as dist

# Charger les modèles Dlib
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")


# Fonction pour calculer l'Eye Aspect Ratio (EAR)
def calculate_ear(eye_points):
    A = dist.euclidean(eye_points[1], eye_points[5])
    B = dist.euclidean(eye_points[2], eye_points[4])
    C = dist.euclidean(eye_points[0], eye_points[3])
    ear = (A + B) / (2.0 * C)
    return ear


# Indices des points des yeux dans les landmarks de Dlib
LEFT_EYE_POINTS = list(range(36, 42))
RIGHT_EYE_POINTS = list(range(42, 48))
EAR_THRESHOLD = 0.25
FRAMES_TO_CONFIRM = 3  # Nombre de frames pour confirmer le clignement

cap = cv2.VideoCapture(0)
blink_counter = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    for face in faces:
        shape = predictor(gray, face)
        shape = np.array([[p.x, p.y] for p in shape.parts()])

        # Calculer l'EAR pour les deux yeux
        left_eye_ear = calculate_ear(shape[LEFT_EYE_POINTS])
        right_eye_ear = calculate_ear(shape[RIGHT_EYE_POINTS])
        ear = (left_eye_ear + right_eye_ear) / 2.0

        if ear < EAR_THRESHOLD:
            blink_counter += 1
        else:
            if blink_counter >= FRAMES_TO_CONFIRM:
                print("Clignement détecté ! Liveness confirmé.")
                blink_counter = 0

    cv2.imshow("Liveness Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# detection des mouvements

```python
import cv2
import dlib
import numpy as np

# Charger les modèles Dlib
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Indices du nez et des yeux
NOSE_INDEX = 30
LEFT_EYE_INDEX = 36
RIGHT_EYE_INDEX = 45
MOVEMENT_THRESHOLD = 15  # Seuil pour considérer un mouvement comme significatif

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    for face in faces:
        shape = predictor(gray, face)
        landmarks = np.array([[p.x, p.y] for p in shape.parts()])

        # Coordonnées du nez et des yeux
        nose_point = landmarks[NOSE_INDEX]
        left_eye = landmarks[LEFT_EYE_INDEX]
        right_eye = landmarks[RIGHT_EYE_INDEX]

        # Calculer la distance horizontale entre le nez et les yeux
        horizontal_movement = abs(left_eye[0] - nose_point[0]) - abs(right_eye[0] - nose_point[0])

        # Déterminer si la tête est tournée
        if horizontal_movement > MOVEMENT_THRESHOLD:
            print("Tête tournée à gauche.")
        elif horizontal_movement < -MOVEMENT_THRESHOLD:
            print("Tête tournée à droite.")
        else:
            print("Tête au centre.")

        # Afficher les points clés
        for (x, y) in landmarks:
            cv2.circle(frame, (x, y), 2, (0, 255, 0), -1)

    cv2.imshow("Détection de mouvement facial", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

## Flux vidéo

```python
import cv2
import dlib
import numpy as np
import mediapipe as mp

# Charger les modèles Dlib
face_rec_model = dlib.face_recognition_model_v1("dlib_face_recognition_resnet_model_v1.dat")
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Initialiser MediaPipe
mp_face_detection = mp.solutions.face_detection

# Ouvrir la webcam (0 pour la webcam par défaut)
cap = cv2.VideoCapture(0)

with mp_face_detection.FaceDetection(min_detection_confidence = 0.5) as face_detection:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Conversion en RGB pour MediaPipe
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = face_detection.process(rgb_frame)

        if results.detections:
            for detection in results.detections:
                # Extraire les coordonnées de la boîte englobante
                bboxC = detection.location_data.relative_bounding_box
                ih, iw, _ = frame.shape
                x = int(bboxC.xmin * iw)
                y = int(bboxC.ymin * ih)
                w = int(bboxC.width * iw)
                h = int(bboxC.height * ih)

                # Dessiner la boîte englobante
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

                # Créer un rectangle compatible Dlib
                rect = dlib.rectangle(x, y, x + w, y + h)

                # Prédire les landmarks avec Dlib
                shape = predictor(frame, rect)

                # Calculer l'encodage du visage avec Dlib
                face_encoding = np.array(face_rec_model.compute_face_descriptor(frame, shape))

                print("Encodage du visage :", face_encoding)

                # Dessiner les landmarks
                for i in range(68):
                    landmark_x = shape.part(i).x
                    landmark_y = shape.part(i).y
                    cv2.circle(frame, (landmark_x, landmark_y), 2, (255, 0, 0), -1)

        # Afficher la vidéo en direct
        cv2.imshow('Webcam', frame)

        # Quitter avec la touche 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

## webcam + nom

```python
import cv2
import dlib
import numpy as np
import mediapipe as mp
from scipy.spatial import distance

# Charger les modèles Dlib
face_rec_model = dlib.face_recognition_model_v1("dlib_face_recognition_resnet_model_v1.dat")
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Initialiser MediaPipe pour la détection de visages
mp_face_detection = mp.solutions.face_detection


# Fonction pour calculer l'encodage d'un visage
def get_face_encoding(image, rect):
    shape = predictor(image, rect)
    encoding = np.array(face_rec_model.compute_face_descriptor(image, shape))
    return encoding


# Fonction pour comparer les encodages
def is_face_known(encoding, known_encodings, threshold=0.6):
    distances = [distance.euclidean(encoding, known) for known in known_encodings]
    return np.min(distances) < threshold, np.min(distances)


# Charger les encodages des visages connus
# Exemple : [{"name": "Personne1", "encoding": array([...])}, ...]
known_faces = [
    {"name": "Alice", "encoding": np.load("alice_encoding.npy")},
    #{"name": "Bob", "encoding": np.load("bob_encoding.npy")}
]

# Ouvrir la webcam
cap = cv2.VideoCapture(0)

with mp_face_detection.FaceDetection(min_detection_confidence = 0.5) as face_detection:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Conversion en RGB pour MediaPipe
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = face_detection.process(rgb_frame)

        if results.detections:
            for detection in results.detections:
                # Extraire les coordonnées de la boîte englobante
                bboxC = detection.location_data.relative_bounding_box
                ih, iw, _ = frame.shape
                x = int(bboxC.xmin * iw)
                y = int(bboxC.ymin * ih)
                w = int(bboxC.width * iw)
                h = int(bboxC.height * ih)

                # Dessiner la boîte englobante
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

                # Créer un rectangle compatible Dlib
                rect = dlib.rectangle(x, y, x + w, y + h)

                # Calculer l'encodage du visage
                face_encoding = get_face_encoding(frame, rect)

                # Vérifier si le visage est connu
                is_known, dist = is_face_known(face_encoding, [f["encoding"] for f in known_faces])
                if is_known:
                    name = [f["name"] for f in known_faces if distance.euclidean(face_encoding, f["encoding"]) == dist][
                        0]
                    cv2.putText(frame, f"{name} ({dist:.2f})", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0),
                                2)
                else:
                    cv2.putText(frame, "Inconnu " + str(dist), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255),
                                2)

        # Afficher le flux vidéo
        cv2.imshow('Reconnaissance faciale', frame)

        # Quitter avec la touche 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
